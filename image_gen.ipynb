```python
from diffusers import DiffusionPipeline
import torch

pipe = DiffusionPipeline.from_pretrained("stabilityai/stable-diffusion-xl-base-1.0", torch_dtype=torch.float16, use_safetensors=True, variant="fp16")
pipe.to("cuda")


prompt = "Create an image of a great Islamic warrior standing tall and majestic, holding a magnificent, ornate blade in one hand. The warrior is adorned in traditional Islamic armor, with intricate details and vivid colors. The background features a dramatic and epic landscape that complements the warrior's heroic presence"

images = pipe(prompt=prompt).images[0]
```
